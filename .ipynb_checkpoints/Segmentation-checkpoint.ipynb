## Segmentation of Nuclei Using Image Processing

Identifying nuclei is the starting point for most cellular analysis because the nucleus contains human DNA. Researchers count the number of cells in a sample by identifying nuclei and measure how cells react to various treatments. In this project I will create an image segmentation to detect the nuclei in cellular images.

Object identification or segmentation is the process of identifying objects within images and separating them from their backgrounds and other irelevant information. A mask is produced to remove the background information and display only the objects.

```python
# list mask image paths
tmp_mask_path = [os.path.join(TRAIN_PATH,tmp_imageid,'masks',mask) for mask in tmp_mask_dir]
```

```python
# find shape
mask0 = cv2.imread(tmp_mask_path[0], cv2.IMREAD_GRAYSCALE)
height = mask0.shape[0]
width = mask0.shape[1]
# make black image
im = np.zeros([height,width],dtype=np.uint8)
# combine masks 
for mask in tmp_mask_path:
    mask = cv2.imread(mask, cv2.IMREAD_GRAYSCALE)
    im = im + mask
tmp_mask = im
# plot
fig, axs = plt.subplots(1, 2, figsize=(10,10))
axs[0].imshow(tmp_image)
axs[1].imshow(tmp_mask)
```

```python
### create function to combine mask images into one mask
def find_mask(dir_path, imageid):
    '''
    dir_path: directory training/ test data
    imageid: the image id to create the mask for
    '''
    tmp_mask_dir = os.listdir(os.path.join(dir_path,imageid,'masks'))
    tmp_mask_path = [os.path.join(dir_path,imageid,'masks',mask) for mask in tmp_mask_dir]
    mask0 = cv2.imread(tmp_mask_path[0], cv2.IMREAD_GRAYSCALE)
    height = mask0.shape[0]
    width = mask0.shape[1]
    im = np.zeros([height,width],dtype=np.uint8)
    for mask in tmp_mask_path:
        mask = cv2.imread(mask, cv2.IMREAD_GRAYSCALE)
        im = im + mask
    return im
```

### Intersection of Union IoU test for accuracy
Intersection of union (IoU) is an evaluation method for object detection, otherwise known as the Jaccard index. This method measures the similarities of a ground truth cell mask to a predicted cell mask. IoU is the area of overlap (the intersection) divided by the area of union. When the IoU is one the prediction is a perfect match with the ground truth mask. 

```python
# intersection
intersection = tmp_mask & thresh
# union
union = tmp_mask | thresh
# symemetric difference
difference = (tmp_mask | thresh) - (tmp_mask & thresh)
fig, axs = plt.subplots(1, 4, figsize=(20,20))
axs[0].imshow(tmp_image) 
axs[1].imshow(tmp_mask) 
axs[2].imshow(intersection)
axs[3].imshow(difference)
```

```python
iou = intersection.sum() / union.sum()
iou
```

```python
### create function for IoU test
def iou_calc(mask_true, mask_pred):
    '''
    mask_true: the true mask
    mask_pred: the predicted mask
    '''
    intersection = mask_true & mask_pred
    union = mask_true | mask_pred
    difference = (mask_true | mask_pred) - (mask_true & mask_pred)
    return intersection.sum() / union.sum()
```

```python
### test function
iou_calc(tmp_mask, thresh)
```

## Evaluate training set
We will evaluate the training set by taking the mean IoU across all the images.

```python
### list of all image ids
imageid_list = list(sorted(set(df_labels['ImageId'])))
```

```python


all_true_masks = []
for id in tqdm(imageid_list):
    mask = find_mask(TRAIN_PATH, id)
    all_true_masks.append(mask)
```

```python
all_pred_masks = []
for id in tqdm(imageid_list):
    tmp_image_path = os.path.join(TRAIN_PATH,id,'images',id+'.png')
    tmp_image = cv2.imread(tmp_image_path, cv2.IMREAD_GRAYSCALE)
    _, thresh = cv2.threshold(tmp_image, 100, 255, cv2.THRESH_OTSU)
    all_pred_masks.append(thresh)
```

```python
all_iou = []
for i in range(len(all_pred_masks)):
    iou = iou_calc(all_true_masks[i],all_pred_masks[i])
    all_iou.append(iou)
```

```python
np.mean(all_iou)
```

### Histogram equalization
There are faint cells blended in with the background that the threshold method does not pick up. We will equalize the image to redistribute the contrast and try to make the faint cells more pronounced.

```python
hist = cv2.calcHist([tmp_image],[0],None,[256],[0,256])
plt.hist(tmp_image.ravel(),256,[0,256])
plt.show()
```

```python
hist.min(),hist.max()
```

```python
# create a CLAHE object (Arguments are optional).
# adaptive histogram equalization
blur = cv2.GaussianBlur(tmp_image,(3,3),0)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(blur)
cl2 = clahe.apply(cl1)
_, thresh_cl2 = cv2.threshold(cl2, 100, 255, cv2.THRESH_OTSU)
fig, axs = plt.subplots(1, 3, figsize=(20,20))
axs[0].imshow(tmp_image)
axs[1].imshow(cl2)
axs[2].imshow(thresh_cl2)
```

```python
# intersection
intersection = tmp_mask & thresh_cl2
# union
union = tmp_mask | thresh_cl2
# symemetric difference
difference = (tmp_mask | thresh_cl2) - (tmp_mask & thresh_cl2)
fig, axs = plt.subplots(1, 4, figsize=(20,20))
axs[0].imshow(tmp_image) 
axs[1].imshow(tmp_mask) 
axs[2].imshow(intersection)
axs[3].imshow(difference)
intersection.sum() / union.sum()
```

```python
all_pred_masks_cl2 = []
for id in tqdm(imageid_list):
    tmp_image_path = os.path.join(TRAIN_PATH,id,'images',id+'.png')
    tmp_image = cv2.imread(tmp_image_path, cv2.IMREAD_GRAYSCALE)
    blur = cv2.GaussianBlur(tmp_image,(3,3),0)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    cl1 = clahe.apply(blur)
    cl2 = clahe.apply(cl1)
    _, thresh_cl2 = cv2.threshold(cl2, 100, 255, cv2.THRESH_OTSU)
    all_pred_masks_cl2.append(thresh_cl2)
```

```python
all_iou_cl2 = []
for i in range(len(all_pred_masks)):
    iou = iou_calc(all_true_masks[i],all_pred_masks_cl2[i])
    all_iou_cl2.append(iou)
np.mean(all_iou_cl2)
```

### Watershed
http://cmm.ensmp.fr/~beucher/wtshed.html

```python
np.random.seed(23)
random_image = df_labels.sample(1)
tmp_imageid = str(random_image.get_value(random_image.index[0],'ImageId'))
tmp_image_path = os.path.join(TRAIN_PATH,tmp_imageid,'images',tmp_imageid+'.png')
tmp_image_path
tmp_image = cv2.imread(tmp_image_path, cv2.IMREAD_GRAYSCALE)
plt.imshow(tmp_image)
```

```python
# noise removal
_, thresh = cv2.threshold(tmp_image, 100, 255, cv2.THRESH_OTSU)
kernel = np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 2)

# identify sure background
sure_bg = cv2.dilate(opening,kernel,iterations=3)

# identify sure foreground area
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5)
ret, sure_fg = cv2.threshold(dist_transform,0.5*dist_transform.max(),255,0)

# identify the unknown region. could be foreground or background
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg,sure_fg)
```

```python
fig, axs = plt.subplots(1, 5, figsize=(20,20))
axs[0].imshow(opening)
axs[1].imshow(sure_bg)
axs[2].imshow(dist_transform)
axs[3].imshow(sure_fg)
axs[4].imshow(unknown)
```

```python
#create markers for watershed
_, markers = cv2.connectedComponents(sure_fg)

# Add one to all labels so that sure background is not 0, but 1
markers = markers+1

# Now, mark the region of unknown with zero
markers[unknown==255] = 0
markers = markers.astype('int32')
```

```python
plt.imshow(markers)
```

```python
tmp_image_3ch = cv2.imread(tmp_image_path, 1)
markers = cv2.watershed(tmp_image_3ch,markers)
tmp_image_3ch[markers == -1] = [255,0,0]
```

```python
tmp_image_3ch = cv2.imread(tmp_image_path, 1)
markers = cv2.watershed(tmp_image_3ch,markers)
#plt.set_cmap('gray')
plt.imshow(markers)
```
